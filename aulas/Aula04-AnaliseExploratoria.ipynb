## 3. Visualizações Avançadas com Matplotlib/Seaborn

### 3.1 Análise Multivariada - Pairplot

**O que é um Pairplot?**
Um pairplot é uma matriz de gráficos que mostra TODAS as relações possíveis entre as variáveis do dataset:
- **Diagonal**: Distribuições individuais de cada variável
- **Off-diagonal**: Relações bivariadas (scatter plots) entre pares de variáveis
- **Cores**: Diferentes categorias/classes para identificar padrões

> 💡 **Por que usar?** Com um único gráfico, você visualiza padrões, correlações e separabilidade entre classes!

````python
# @title
# Pairplot avançado com customizações
g = sns.pairplot(iris,
                hue='species',
                height=2.5,
                aspect=1.1,
                markers=['o', 's', 'D'],
                diag_kind='kde',
                plot_kws={'alpha': 0.7, 'edgecolor': 'white', 'linewidth': 0.5},
                diag_kws={'alpha': 0.7, 'linewidth': 2})

g.fig.suptitle('Análise Multivariada: Todas as Relações em Uma Visualização',
              y=1.02, size=16, weight='bold')

# Adicionando anotações
g.axes[0, 0].annotate('Distribuições\nna diagonal',
                     xy=(0.5, 0.95),
                     xycoords='axes fraction',
                     ha='center',
                     fontsize=9,
                     bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.3))

plt.show()

print("🎯 INSIGHTS DO PAIRPLOT:")
print("• Diagonal: Distribuições de cada característica por espécie")
print("• Off-diagonal: Relações bivariadas entre características")
print("• Setosa (círculos): Claramente separada das outras espécies")
print("• Versicolor (quadrados) e Virginica (diamantes): Alguma sobreposição")
print("• Características da pétala: Melhor separação entre espécies")
````

### 3.2 Heatmap de Correlação Avançado

**O que é um Heatmap de Correlação?**
Um heatmap é uma representação visual de uma matriz de correlação onde:
- **Cores quentes (vermelho)**: Correlação positiva forte (variáveis crescem juntas)
- **Cores frias (azul)**: Correlação negativa forte (uma cresce, outra diminui)
- **Cores neutras (branco)**: Sem correlação
- **Valores**: Coeficientes de correlação de Pearson (-1 a +1)

> 🔍 **Insight importante**: Vamos comparar correlações gerais vs. por espécie para descobrir padrões ocultos!

````python
# @title
# Calculando correlações por espécie
fig, axes = plt.subplots(1, 4, figsize=(18, 4))

# Correlação geral
corr_all = iris[features].corr()
sns.heatmap(corr_all, annot=True, cmap='RdBu_r', center=0,
           vmin=-1, vmax=1, square=True, ax=axes[0],
           cbar_kws={"shrink": 0.8})
axes[0].set_title('Todas as Espécies', fontsize=12, weight='bold')

# Correlação por espécie
for idx, (species, ax) in enumerate(zip(iris['species'].unique(), axes[1:])):
    subset = iris[iris['species'] == species]
    corr = subset[features].corr()
    sns.heatmap(corr, annot=True, cmap='RdBu_r', center=0,
               vmin=-1, vmax=1, square=True, ax=ax,
               cbar_kws={"shrink": 0.8})
    ax.set_title(f'{species.capitalize()}', fontsize=12, weight='bold')

plt.suptitle('Padrões de Correlação: Geral vs. Por Espécie',
            fontsize=14, weight='bold', y=1.05)
plt.tight_layout()
plt.show()

print("🔍 DESCOBERTA IMPORTANTE:")
print("As correlações variam entre espécies! Isso sugere que:")
print("• Cada espécie tem padrões de crescimento diferentes")
print("• Modelos específicos por espécie podem ser mais eficazes")
print("• A separação entre espécies não é apenas por tamanho")
````

### 3.3 Análise de Clusters com KDE 2D

**O que é KDE (Kernel Density Estimation)?**
KDE é uma técnica que estima a densidade de probabilidade dos dados:
- **Contornos de densidade**: Mostram onde os dados estão mais concentrados
- **Clusters**: Grupos naturais de dados com alta densidade
- **Sobreposição**: Áreas onde diferentes grupos se misturam

> 🎯 **Objetivo**: Identificar visualmente quais combinações de variáveis melhor separam as espécies e onde ocorrem sobreposições!

````python
# @title
# Visualização de densidade 2D para identificar clusters
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

# Combinações de variáveis mais informativas
combinations = [
    ('petal_length', 'petal_width', 'Melhor Separação'),
    ('sepal_length', 'petal_length', 'Segunda Melhor'),
    ('sepal_length', 'sepal_width', 'Mais Sobreposição'),
    ('petal_length', 'sepal_width', 'Padrão Interessante'),
    ('petal_width', 'sepal_width', 'Contraste Alto'),
    ('sepal_length', 'petal_width', 'Visão Alternativa')
]

for idx, (x_var, y_var, title) in enumerate(combinations):
    ax = axes[idx]

    # Scatter plot com cores
    for species in iris['species'].unique():
        subset = iris[iris['species'] == species]
        ax.scatter(subset[x_var], subset[y_var],
                  label=species, alpha=0.6, s=50, edgecolors='white')

    # Adicionando contornos de densidade
    for species in iris['species'].unique():
        subset = iris[iris['species'] == species]
        sns.kdeplot(data=subset, x=x_var, y=y_var, ax=ax,
                   levels=2, alpha=0.3, linewidths=1)

    ax.set_xlabel(x_var.replace('_', ' ').title())
    ax.set_ylabel(y_var.replace('_', ' ').title())
    ax.set_title(title, fontsize=11, weight='bold')
    ax.legend(loc='best', fontsize=8)
    ax.grid(True, alpha=0.3)

plt.suptitle('Análise de Clusters: Diferentes Perspectivas',
            fontsize=14, weight='bold', y=1.02)
plt.tight_layout()
plt.show()

print("🎯 ANÁLISE DE CLUSTERS:")
print("• Petal length × width: 3 clusters bem definidos")
print("• Sepal length × width: Alta sobreposição, difícil separação")
print("• Contornos de densidade revelam a forma dos clusters")
print("• Setosa forma clusters compactos e isolados")
````

## 4. Identificação de Padrões

### 4.1 Análise de Componentes Principais Simplificada

> 💡 **O que é PCA?**
> PCA (Principal Component Analysis) é uma técnica que reduz a dimensionalidade dos dados:
> - **Objetivo**: Transformar muitas variáveis em poucas, mantendo a maior parte da informação
> - **Como funciona**: Encontra direções (componentes) que capturam a maior variação nos dados
> - **Analogia**: É como tirar uma foto 2D de um objeto 3D - você perde alguma informação, mas escolhe o melhor ângulo
> - **Neste caso**: Reduzimos 4 dimensões (características) para 2 dimensões (componentes)

````python
# @title
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Padronizando os dados
scaler = StandardScaler()
X_scaled = scaler.fit_transform(iris[features])

# Aplicando PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Criando DataFrame com componentes principais
pca_df = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
pca_df['species'] = iris['species'].values

# Visualização
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Subplot 1: PCA scatter
ax1 = axes[0]
for species in iris['species'].unique():
    subset = pca_df[pca_df['species'] == species]
    ax1.scatter(subset['PC1'], subset['PC2'],
               label=species, alpha=0.7, s=100, edgecolors='white')

ax1.set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.1%} da variância)')
ax1.set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.1%} da variância)')
ax1.set_title('Redução de Dimensionalidade: 4D → 2D', fontsize=12, weight='bold')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Subplot 2: Importância das características
ax2 = axes[1]
components = pd.DataFrame(
    pca.components_.T,
    columns=['PC1', 'PC2'],
    index=features
)
components_abs = components.abs()

# Criando o gráfico de barras manualmente
x = np.arange(len(features))
width = 0.35

bars1 = ax2.bar(x - width/2, components_abs['PC1'], width, label='PC1', color='steelblue')
bars2 = ax2.bar(x + width/2, components_abs['PC2'], width, label='PC2', color='coral')

ax2.set_xlabel('Características')
ax2.set_ylabel('Importância (valor absoluto)')
ax2.set_title('Contribuição das Características', fontsize=12, weight='bold')
ax2.set_xticks(x)
ax2.set_xticklabels([f.replace('_', '\n') for f in features], rotation=45)
ax2.legend(title='Componente')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"📊 ANÁLISE DE PADRÕES:")
print(f"• Variância explicada: {sum(pca.explained_variance_ratio_):.1%} com apenas 2 componentes")
print(f"• PC1 captura principalmente variação no tamanho (pétalas dominam)")
print(f"• PC2 captura contraste entre sépalas e pétalas")
print(f"• Separação clara entre espécies mesmo em 2D")

# Mostrando os valores exatos das contribuições
print(f"\n📈 Contribuições das características:")
for feature in features:
    print(f"  {feature}:")
    print(f"    PC1: {abs(components.loc[feature, 'PC1']):.3f}")
    print(f"    PC2: {abs(components.loc[feature, 'PC2']):.3f}")
````

## 5. Visualizações 3D Interativas com Plotly

**Por que visualizações 3D interativas?**
Visualizações 3D permitem explorar relações complexas entre múltiplas variáveis:
- **3 eixos (X, Y, Z)**: Três variáveis numéricas simultaneamente
- **Cor**: Quarta dimensão (categoria/classe)
- **Tamanho**: Quinta dimensão (outra variável numérica)
- **Interatividade**: Rotação, zoom e hover para exploração detalhada

> 🌐 **Vantagem**: Em uma única visualização, podemos representar até 5 dimensões de informação!

````python
# @title
# Visualização 3D interativa
fig = px.scatter_3d(iris,
                   x='sepal_length',
                   y='sepal_width',
                   z='petal_length',
                   color='species',
                   size='petal_width',
                   hover_data={'petal_width': ':.2f'},
                   title='Visualização 3D Interativa: 4 Dimensões em um Gráfico',
                   labels={'sepal_length': 'Comprimento Sépala',
                          'sepal_width': 'Largura Sépala',
                          'petal_length': 'Comprimento Pétala'},
                   opacity=0.8)

fig.update_layout(
    scene=dict(
        xaxis_title='Comprimento da Sépala (cm)',
        yaxis_title='Largura da Sépala (cm)',
        zaxis_title='Comprimento da Pétala (cm)',
        camera=dict(
            eye=dict(x=1.5, y=1.5, z=1.5)
        )
    ),
    width=900,
    height=600
)

fig.show()

print("🌐 VANTAGENS DA VISUALIZAÇÃO 3D INTERATIVA:")
print("• Rotação: Explore diferentes ângulos arrastando o mouse")
print("• Zoom: Use scroll para aproximar/afastar")
print("• 4ª dimensão: Tamanho dos pontos = largura da pétala")
print("• Hover: Passe o mouse para ver valores exatos")
print("• Isolamento: Clique na legenda para mostrar/ocultar espécies")
````

````python
# @title
# Dashboard interativo com Plotly
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=('Distribuição: Comprimento da Pétala',
                   'Boxplot: Largura da Pétala por Espécie',
                   'Scatter: Relação Pétala',
                   'Radar: Perfil Médio por Espécie'),
    specs=[[{'type': 'histogram'}, {'type': 'box'}],
           [{'type': 'scatter'}, {'type': 'polar'}]]
)

# 1. Histograma
for species in iris['species'].unique():
    subset = iris[iris['species'] == species]
    fig.add_trace(
        go.Histogram(x=subset['petal_length'], name=species, opacity=0.7),
        row=1, col=1
    )

# 2. Boxplot
for species in iris['species'].unique():
    subset = iris[iris['species'] == species]
    fig.add_trace(
        go.Box(y=subset['petal_width'], name=species),
        row=1, col=2
    )

# 3. Scatter
for species in iris['species'].unique():
    subset = iris[iris['species'] == species]
    fig.add_trace(
        go.Scatter(x=subset['petal_length'], y=subset['petal_width'],
                  mode='markers', name=species,
                  marker=dict(size=8, opacity=0.7)),
        row=2, col=1
    )

# 4. Radar Chart
categories = features
for species in iris['species'].unique():
    subset = iris[iris['species'] == species]
    values = subset[features].mean().values.tolist()
    values.append(values[0])  # Fechar o polígono

    fig.add_trace(
        go.Scatterpolar(
            r=values,
            theta=categories + [categories[0]],
            fill='toself',
            name=species,
            opacity=0.6
        ),
        row=2, col=2
    )

# Atualizando layout
fig.update_layout(
    height=700,
    showlegend=True,
    title_text="Dashboard Interativo: Múltiplas Perspectivas",
    title_font_size=16
)

# Ajustes específicos
fig.update_xaxes(title_text="Comprimento da Pétala (cm)", row=1, col=1)
fig.update_xaxes(title_text="Comprimento da Pétala (cm)", row=2, col=1)
fig.update_yaxes(title_text="Frequência", row=1, col=1)
fig.update_yaxes(title_text="Largura da Pétala (cm)", row=1, col=2)
fig.update_yaxes(title_text="Largura da Pétala (cm)", row=2, col=1)

fig.show()

print("📊 DASHBOARD INSIGHTS:")
print("• Histograma: Distribuições bimodais indicam separação natural")
print("• Boxplot: Setosa tem menor variabilidade")
print("• Scatter: Correlação linear forte entre dimensões da pétala")
print("• Radar: Perfis distintos - Setosa é menor em todas as dimensões")
````
